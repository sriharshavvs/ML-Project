```python
import matplotlib.pyplot as plt
import numpy as np

# Define the performance metrics for R (replace with your actual values)
regressors = ['Linear Regressor', 'Decision Tree', 'Voting Regressor', 'XGBoost', 'CatBoost']
techniques = ['Default', 'PCA', 'UMAP', 't-SNE']

# Example R values for each technique
r2_scores = {
    'Default': [-4.72, 0.43, -0.69, 0.63, 0.70],
    'PCA': [0.23, -0.01, 0.3, 0.47, 0.57],
    'UMAP': [-0.01, 0.06, 0.28, 0.37, 0.45],
    't-SNE': [0, -0.36, 0.08, 0.06, 0.15]
}

# Plot settings
x = np.arange(len(regressors))  # Regressor indices
width = 0.2  # Width of bars

fig, ax = plt.subplots(figsize=(12, 6))

# Add bars for each technique
for i, technique in enumerate(techniques):
    ax.bar(x + i * width, r2_scores[technique], width, label=technique)

# Customize the plot
ax.set_xlabel('Regressors', fontsize=12)
ax.set_ylabel('$R^2$ Score', fontsize=12)
ax.set_title('Comparison of R Values Across Dimensionality Reduction Techniques', fontsize=14)
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(regressors, rotation=45, ha='right', fontsize=10)
ax.legend(title='Reduction Technique', fontsize=10)
ax.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Add a line at R=0
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

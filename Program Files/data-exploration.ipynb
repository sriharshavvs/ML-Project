```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define file paths for all datasets
file_paths = [
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_cq_embed_data.csv",
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_cs_embed_data.csv",
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_cc_embed_data.csv",
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_co_embed_data.csv"
]

# Combine all datasets into one DataFrame
combined_df = pd.concat([pd.read_csv(file_path) for file_path in file_paths], ignore_index=True)

# Display basic information about the combined dataset
print("Basic Information:")
print(combined_df.info())

# Display first 5 rows
print("\nFirst 5 Rows:")
print(combined_df.head())

# Check for missing values
missing_values = combined_df.isnull().sum()
print("\nMissing Values:")
print(missing_values)

# Summary statistics for numeric columns
print("\nSummary Statistics:")
print(combined_df.describe())

# 1. Histogram of 'Final_Marks'
plt.figure(figsize=(8, 6))
combined_df['Final_Marks'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Final Marks")
plt.xlabel('Final Marks')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 2. Boxplot of 'Final_Marks'
plt.figure(figsize=(8, 6))
sns.boxplot(x=combined_df['Final_Marks'], color='lightgreen')
plt.title("Boxplot of Final Marks")
plt.xlabel('Final Marks')
plt.tight_layout()
plt.show()

# 3. Count plot for 'error_count' (assuming 'error_count' exists as a categorical column)
if 'error_count' in combined_df.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='error_count', data=combined_df, palette='Set2')
    plt.title("Count Plot of Error Count")
    plt.xlabel('Error Count')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# 4. Pairplot for relationships between numerical columns
numeric_columns = combined_df.select_dtypes(include=['number']).columns
plt.figure(figsize=(8, 6))
sns.pairplot(combined_df[numeric_columns])
plt.title("Pairplot of Numerical Columns")
plt.tight_layout()
plt.show()

# 5. Heatmap of correlations between numerical columns
plt.figure(figsize=(10, 8))
correlation_matrix = combined_df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Columns")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
file_paths = [
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_cq_embed_data.csv",
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_cs_embed_data.csv",
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_cc_embed_data.csv",
    r"C:\Users\vvmad\Downloads\5th\ML\PROJECT\datasets\java_co_embed_data.csv"
]

# Initialize an empty list to collect 'final_marks' from all datasets
combined_final_marks = []

# Process each file and collect 'final_marks'
for file_path in file_paths:
    # Load dataset
    df = pd.read_csv(file_path)
    
    # Check if 'final_marks' column exists
    if 'Final_Marks' in df.columns:
        combined_final_marks.extend(df['Final_Marks'].values)
    else:
        print(f"'final_marks' column not found in {file_path}.")

# Create a DataFrame for the combined final_marks
combined_df = pd.DataFrame(combined_final_marks, columns=['Final_Marks'])

# Calculate frequency of combined final_marks
frequency = combined_df['Final_Marks'].value_counts()

# Plot the frequency distribution of the combined final_marks
plt.figure(figsize=(10, 6))
frequency.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Frequency Distribution of 'final_marks' (Combined Data)")
plt.xlabel('Final Marks')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
